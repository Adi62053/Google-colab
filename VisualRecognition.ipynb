```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image (change name if needed)
image = cv2.imread('img5.webp')

# Safety check
if image is None:
    raise ValueError("Image not loaded. Check filename!")

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Fixed threshold value
T = 127

# Manual thresholding using NumPy (NOT built-in)
binary = np.zeros_like(gray)
binary[gray >= T] = 255

# Display results
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Grayscale Image")
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Binary Output (Manual Threshold)")
plt.imshow(binary, cmap='gray')
plt.axis('off')

plt.show()
```

```python
# ================================
# IMPORT REQUIRED LIBRARIES
# ================================
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# ================================
# UPLOAD IMAGE FROM YOUR SYSTEM
# ================================
print("Upload an image from your system")
uploaded = files.upload()

# Read the uploaded image
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# ================================
# TASK 1: EDGE DETECTION METHODS
# ================================

# ---- Sobel Edge Detection ----
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
sobel = cv2.magnitude(sobel_x, sobel_y)

# ---- Prewitt Edge Detection ----
prewitt_x = np.array([[1, 0, -1],
                      [1, 0, -1],
                      [1, 0, -1]])

prewitt_y = np.array([[1,  1,  1],
                      [0,  0,  0],
                      [-1, -1, -1]])

px = cv2.filter2D(gray, -1, prewitt_x)
py = cv2.filter2D(gray, -1, prewitt_y)
prewitt = cv2.magnitude(px.astype(float), py.astype(float))

# ---- Laplacian Edge Detection ----
laplacian = cv2.Laplacian(gray, cv2.CV_64F)

# ================================
# DISPLAY TASK 1 OUTPUTS
# ================================
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title("Original Grayscale Image")
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Sobel Edge Detection")
plt.imshow(sobel, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("Prewitt Edge Detection")
plt.imshow(prewitt, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Laplacian Edge Detection")
plt.imshow(laplacian, cmap='gray')
plt.axis('off')

plt.show()

# ================================
# TASK 2: CANNY EDGE DETECTION
# ================================

# ---- Effect of Gaussian Blurring ----
blur_3 = cv2.GaussianBlur(gray, (3,3), 0)
blur_7 = cv2.GaussianBlur(gray, (7,7), 0)
blur_11 = cv2.GaussianBlur(gray, (11,11), 0)

canny_3 = cv2.Canny(blur_3, 50, 150)
canny_7 = cv2.Canny(blur_7, 50, 150)
canny_11 = cv2.Canny(blur_11, 50, 150)

# ================================
# DISPLAY GAUSSIAN BLUR EFFECT
# ================================
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title("Canny with 3x3 Gaussian Blur")
plt.imshow(canny_3, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Canny with 7x7 Gaussian Blur")
plt.imshow(canny_7, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("Canny with 11x11 Gaussian Blur")
plt.imshow(canny_11, cmap='gray')
plt.axis('off')

plt.show()

# ================================
# EFFECT OF HYSTERESIS THRESHOLDS
# ================================
canny_low = cv2.Canny(gray, 30, 100)
canny_medium = cv2.Canny(gray, 50, 150)
canny_high = cv2.Canny(gray, 100, 200)

# ================================
# DISPLAY THRESHOLD EFFECT
# ================================
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title("Low Thresholds (30, 100)")
plt.imshow(canny_low, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Medium Thresholds (50, 150)")
plt.imshow(canny_medium, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("High Thresholds (100, 200)")
plt.imshow(canny_high, cmap='gray')
plt.axis('off')

plt.show()
```
