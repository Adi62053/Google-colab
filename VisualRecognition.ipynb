```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image (change name if needed)
image = cv2.imread('img5.webp')

# Safety check
if image is None:
    raise ValueError("Image not loaded. Check filename!")

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Fixed threshold value
T = 127

# Manual thresholding using NumPy (NOT built-in)
binary = np.zeros_like(gray)
binary[gray >= T] = 255

# Display results
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Grayscale Image")
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Binary Output (Manual Threshold)")
plt.imshow(binary, cmap='gray')
plt.axis('off')

plt.show()
```
